# Parsing an example text

```python
import amrlib
stog = amrlib.load_stog_model()
graphs = stog.parse_sents(['Napoleon Bonaparte, later known by his regnal name Napoleon I, was a French military commander and political leader who rose to prominence during the French Revolution and led successful campaigns during the Revolutionary Wars.'],add_metadata=False)
for graph in graphs:
    print(graph)
```

Importing penamn libraries, we can extract the triples from the graph

```python
import penman
from penman import layout

def extract_triples(amr_sentence):
    
    amr_decoded = penman.decode(amr_sentence)
    
    triples = []
    for triple in amr_decoded.triples:
        triples.append(triple)
    return triples

triples = extract_triples(graphs[0])
for triple in triples:
    print(triple)
```

## Let's use networkx and put that in a graph 
with the help of Networkx


```python
import networkx as nx
import re

def add_triples_to_graph(triples):
    graph = nx.DiGraph()

    for triple in triples:
        source, relation, target = triple
        graph.add_node(source, instance=None)
        graph.add_node(target, instance=None)
        graph.add_edge(source, target, label=relation)

    return graph

nx_graph = add_triples_to_graph(triples)

# Print the nodes and edges in the graph
print("Nodes:", nx_graph.nodes())
print("Edges:")
for edge in nx_graph.edges(data=True):
    print(edge[0], "--", edge[2]["label"], "-->", edge[1])
    
```

Proceed to Visualization

```python
import matplotlib.pyplot as plt

def visualize_graph(graph):
    pos = nx.spring_layout(graph, seed=42)  # Position the nodes using the spring layout algorithm
    labels = nx.get_edge_attributes(graph, 'label')
    labels = {k: v for k, v in labels.items()}

    # Draw nodes and edges
    nx.draw(graph, pos, with_labels=True, node_size=800, node_color='lightblue', font_size=8, font_color='black')
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=labels, font_size=8, label_pos=0.3)

    plt.axis('off')
    plt.show()

visualize_graph(nx_graph)
```

```python
from pyvis.network import Network
from IPython.display import display, HTML

def visualize_3d_graph(graph):
    
    nt = Network(notebook=True, cdn_resources='remote')
    # populates from the Networkx graph created earlier
    nt.from_nx(graph)
    nt.show('nx.html')

visualize_3d_graph(nx_graph)
```

Now that we have the full articles, we will proceed to extract the introduction from each by creating a function

```python
def extract_first_part(article_text):
    
    """
    Extracts the first part of an article up to the second double line break.
    """
    
    second_line_break = article_text.find('\n\n', article_text.find('\n\n') + 1)
    return article_text[:second_line_break] if second_line_break != -1 else article_text
```

Let's see how it works.

```python
article_text = wikipedia_dataset[70]['text']
first_part = extract_first_part(article_text)

print(f"Extracted First Part:\n{first_part}")
```

## Now that we have the introduction, we can build a function that takes as input an article and return the AMR Graph 

```python
import amrlib

# Lets turn that into a function to reuse while the dataset downloads:

def article__to_amr(article):
    
    """Takes as input text of Wikipedia English dataset article: input as wikipedia_dataset[index]['text']
        Returns a list of AMRs per article"""
    
    # Inject the sentence to AMR graph model installed 
    stog = amrlib.load_stog_model()
    
    # Extract Introduction 
    text_content = extract_first_part(article)

    # Check if the string is not empty, proceeds to         
    # Parse the text content and add the parsed graph to the list
    if text_content:
        graphs = stog.parse_sents([text_content], add_metadata=False) 
        # Square brackets because .parse_sents takes list of sentences for input
        # So we wrap the string in a list
    
    return graphs
 
```

```python
article = wikipedia_dataset[10]['text']
graph = article__to_amr(article)
for graph in graphs:
    print(graph)
```

## Now, we can go modify this once again, to make it work on multiple articles.

To do so, we can leverage the amrlib model.parse_sents, which takes as input a list of strings.
We will have as our list of strings, the introduction parts.
For input, we want to be able to pass multiple articles. 

```python
def articles_to_amr(start_index, end_index): 
    
    """
    Takes indices (start_index and end_index) to specify a slice of the Wikipedia English dataset.
    Returns a list of AMRs for the selected articles.
    """
    
    # Inject the sentence to AMR graph model installed 
    stog = amrlib.load_stog_model()
    
    # Gets introduction from articles using the extractor created earlier
    list_of_intros = [extract_first_part(wikipedia_dataset[i]['text']) for i in range (start_index, end_index)]
           
    # Parse the content at once 
    graphs = stog.parse_sents(list_of_intros, add_metadata=False) 
    # Notice how we dont need square brackets aymore
    return graphs
 
```

To split a paragrah into sentences, we can leverage nltk and its method. To make my code more readable, I am putting it into methods I have defined. 

```python
import nltk

def tokenize_into_sentences(text):
    
    """
    Tokenize a text into sentences using NLTK.
    """
    sentences = nltk.sent_tokenize(text)
    return sentences
```

## Lets try and define an article  info extractor

```python
import re 

def extract_articles_info(wikitext, max_articles=None):
    # Initialize variables for titles and contents
    titles = []
    contents = []
    article_counter = 0

    # Initialize variables for the current article
    content = ''
    in_article = False  # Flag to indicate whether currently processing article content

    # Iterate through the dataset
    for row in range(len(wikitext)):
        line = wikitext[row]["text"].strip()
    
        #if line.startswith('='):
        if re.match(r'^= [^=].*?=$', line) and not in_article:
            # If the line starts with exactly one '= ', it's a title
            titles.append(wikitext[row]["text"].strip())
            content = ''
            in_article = True
            article_counter += 1
        elif line.startswith('= = '):
            # If the line starts with exactly two '==', it's the end of the article
            if in_article:
                contents.append(content.strip())
                in_article = False
        
                if max_articles and article_counter >= max_articles:
                    break
        elif in_article:
            # If in_article is True, it means we are still processing article content
            content += wikitext[row]["text"].strip()

    return titles, contents, article_counter 
```

```python
titles, contents, count= extract_articles_info(wikitext_raw)
```

## Parse sentence level and Paragraph lever, and pickle the results for easy access

```python
import pickle

# Initialize storage lists
paragraph_level_graphs = []
sentence_level_graphs  = []  #This SHOULD be a list of lists, accessing the article with first index and sentence with second. It probably wont work in Vulcan. 
for article in contents:

    # Uses NLTK wrapped method to extract sentence from the paragraph/article.
    article_sents = tokenize_into_sentences(article)

    # Parses paragraph level
    paragraph_level_graphs.append(stog.parse_sents([article], add_metadata=False))

    # Parses sentence level 
    sentence_level_graphs.append(stog.parse_sents(article_sents, add_metadata=False))

```

```python
with open('paragraph_level_graphs.pk1', "wb") as file1:
    pickle.dump(paragraph_level_graphs, file1)
    
with open('sentence_level_graphs.pk1', "wb") as file2:
    pickle.dump(sentence_level_graphs, file2)
    
```

The following cell write the AMRs to text files in a similar fashion as the pickled version. However, the sentence level parsing will lose some information, namely article divides. (Essesntialy, an AMR per sentence for the whole dataset)

```python
## Double loop to extract sentences from articles. Here is where I am losing article devides...
with open('sentences.txt', "w") as file:
    for article in sentence_level_graphs:
        for sentence in article:
            file.write(sentence + "\n")
with open('paragraph.txt', "w") as file:
    for article in paragraph_level_graphs:
        file.write(article[0] + "\n")
```
